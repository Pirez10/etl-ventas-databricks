```python
# Leer datos desde capa bronze
ventas_raw = spark.table("bronze.ventas_raw")
productos = spark.table("bronze.productos")

log("Datos de entrada cargados desde capa bronze")

# -------------------------
# 1. Limpieza de datos
# -------------------------
ventas_clean = limpiar_ventas(ventas_raw)
validar_df(ventas_clean, "ventas_clean")

# -------------------------
# 2. Enriquecimiento (join con productos)
# -------------------------
ventas_enriquecidas = enriquecer_ventas(ventas_clean, productos)

# -------------------------
# 3. Agregaciones y métricas
# -------------------------
resumen_diario = calcular_resumen(ventas_enriquecidas)

validar_df(resumen_diario, "resumen_diario")

# -------------------------
# 4. Guardar en capa silver
# -------------------------
resumen_diario.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("silver.resumen_diario")

log("Transform completo - datos guardados en capa silver")
```
